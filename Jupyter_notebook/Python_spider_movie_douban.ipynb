```python
# 抓取数据
dbmovieurl = 'https://movie.douban.com/cinema/nowplaying/guangzhou/'
HEADERS = {
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.26 Safari/537.36 Core/1.63.5377.400 QQBrowser/10.1.1335.400',
    'Referer':'https://movie.douban.com/'
} # 进行‘User-Agetn’和‘Referer’的设定
mr = requests.get(dbmovieurl,headers = HEADERS)
# print(mr.content.decode('utf-8')) 返回的是原生的字符串就是从网页上抓取下来的，没有经过处理的，是bytes类型，解码成‘utf-8’
# print(mr.text) # 返回的是Str(unicode)类型
text = mr.text
print(type(text)) # <class 'str'>
```

```python
# 进行解析
from lxml import etree
html = etree.HTML(text)
print(html)
ul = html.xpath("//ul[@class = 'lists']")[0]
# print(ul) # 拿到正在上映的电影 ul
# print(etree.tostring(ul,encoding = 'utf-8').decode('utf-8')) #对ul进行解析到str
# titles = html.xpath("//ul[@class = 'lists']/li/@data-title")
# print(titles)
# pic = html.xpath("//li[@class = 'poster']/a/@href")
# print(pic)
lis = ul.xpath("./li")
movies=[]
for li in lis:
    title = li.xpath("@data-title")[0]
    score = li.xpath("@data-score")[0]
    duration = li.xpath("@data-duration")[0]
    region = li.xpath("@data-region")[0]
    director = li.xpath("@data-director")[0]
    actors = li.xpath("@data-actors")[0]
    thumbnail = li.xpath(".//img/@src")[0]
    movie = {
        'title':title,
        'score':score,
        'duration':duration,
        'region':region,
        'director':director,
        'actors':actors,
        'thumbnail':thumbnail
    }
    movies.append(movie)
print(movies)
    # print(etree.tostring(li,encoding='utf-8').decode('utf-8'))
```

```python
import requests,re
HEADERS = {
    'User-Agetn':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.26 Safari/537.36 Core/1.63.5377.400 QQBrowser/10.1.1335.400',
    'Referer':'https://www.gushiwen.org/'
}
def parse_url(url):
    res = requests.get(url,headers = HEADERS)
    text = res.text
#    print(res.text)
    titles = re.findall(r'<div\sclass="cont">.*?<b>(.*?)</b>',text,re.DOTALL)
    dynasties = re.findall(r'<p\sclass="source">.*?<a.*?>(.*?)</a>',text,re.DOTALL)
    authors = re.findall(r'<p\sclass="source">.*?</span><a.*?>(.*?)</a>',text,re.DOTALL)
    contents_tags = re.findall(r'<div\sclass="contson".*?>(.*?)</div>',text,re.DOTALL)
    contents=[]
    for content in contents_tags:
        x = re.sub(r'<.*?>',"",content)
        contents.append(x.strip())
        #print(x.strip())
    # print(titles,dynasties,authors,poems)
    gs = zip(titles,dynasties,authors,contents)
    #print(list(gs))
    poems = []
    for value in gs:
        title,dynasty,author,content = value
        poem = {
            'title':title,
            'dynasty':dynasty,
            'author':author,
            'content':content    
        }
        poems.append(poem)
    for poem in poems: 
        print(poem)
        print('.'*40)
        
def main():
    for i in range(1,101):
        url = 'https://www.gushiwen.org/default_{}.aspx'.format(i)
        parse_url(url)
        print('-'*40,"这是第{}页".format(i),'-'*40)
if __name__ == '__main__':
    main()
```

```python
import requests,re
HEADERS = {
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.26 Safari/537.36 Core/1.63.5377.400 QQBrowser/10.1.1335.400',
    'Referer':'http://www.qiushibaike.com/'
}
def parse_html(url):
    res = requests.get(url,headers = HEADERS)
    text = res.text
    author_imgs = re.findall(r'(?<=\bsrc=)\w+\b',text,re.DOTALL)
    print(author_imgs)
def main():
    url = 'https://www.qiushibaike.com/8hr/page/1/'
    parse_html(url)

if __name__ == '__main__':
    main()
```
