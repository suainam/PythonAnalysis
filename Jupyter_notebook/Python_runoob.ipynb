isinstance 和 type 的区别 :
+ type()不会认为子类是一种父类类型；
+ isinstance()会认为子类是一种父类类型

```python
class A:
    pass

class B(A):
    pass

isinstance(A(), A)
```

```python
type(A()) == A
```

```python
isinstance(B(), A)
```

```python
type(B()) == A
```

```python
type(A()), type(B())
```
